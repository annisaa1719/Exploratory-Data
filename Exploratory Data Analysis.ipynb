Iris Dataset: Classification Models

Import Modules

```python
from sklearn.model_selection import train_test_split #pembagi dataset menjadi training dan testing test
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report #evaluasi performa model
```

Dataset: Features & Class Label

```python
X = iris_df.drop(columns='Species') # menempatkan features ke dalam variable X
X.head() # tampilkan 5 baris pertama
```

```python
y = iris_df['Species'] # menempatkan class label (target) ke dalam variable y
y.head() # tampilkan 5 baris pertama
```

Split the dataset into a training set and a testing set

```python
# membagi dataset ke dalam training  dan testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10) #test_size--> testing dataset 40%, training 60%

print('training dataset')
print(X_train.shape)
print(y_train.shape)
print()
print('testing dataset:')
print(X_test.shape)
print(y_test.shape)
```

K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_range = list(range(1,26)) #penjajakan
scores = []
for k in k_range:
    model_knn = KNeighborsClassifier(n_neighbors=k) # konfigurasi algoritma, k--> parameter neighbors
    model_knn.fit(X_train, y_train) # training model/Classifier
    y_pred = model_knn.predict(X_test) # melakukan prediksi
    scores.append(accuracy_score(y_test, y_pred)) # evaluasi performa, nilai dari hasil prediksi, membandingkan nilai sesungguhnya dg nilai prediksi
```

```python
plt.plot(k_range, scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of k of k-Nearest Neighbors')
plt.tight_layout()
plt.show()
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3) # konfigurasi algoritma
model_knn.fit(X_train,y_train) # training model/classifier
y_pred = model_knn.predict(X_test) # melakukan prediksi
```

Accuracy Score

```python
print(accuracy_score(y_test, y_pred)) #evaluasi akurasi
```

Confusion Matrix

```python
print(confusion_matrix(y_test, y_pred)) # evaluasi confusin matrix
```

Classification Report

```python
print(classification_report(y_test, y_pred)) # evaluasi klasifikasi
```
